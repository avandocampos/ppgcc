```python
from utils import *

# Carregar os dados
X, y = load_data("ocr_car_numbers_rotulado.txt")

# Verificar as dimensões dos dados
print("Dimensões dos dados:", X.shape)
print("Dimensões dos rótulos:", y.shape)

# Redimensionar os dados para imagens 35x35
X = X.reshape(-1, 35, 35)

print("Dimensões das imagens:", X.shape)
```

```python
from extractors import *

hog_features = extract_hog_features(X)
print("Dimensões das características HOG:", hog_features.shape)

lbp_features = extract_lbp_features(X)
print("Dimensões das características LBP:", lbp_features.shape)

pca_features = extract_pca_features(X)
print("Dimensões das características PCA:", pca_features.shape)
```

```python
from classifiers import *

# Avaliar o SVM com características HOG
acc_hog, cm_hog = train_evaluate_svm(hog_features, y)
print("Acurácia SVM com HOG:", acc_hog)
print("Matriz de Confusão SVM com HOG:\n", cm_hog)

# Avaliar o SVM com características LBP
acc_lbp, cm_lbp = train_evaluate_svm(lbp_features, y)
print("Acurácia SVM com LBP:", acc_lbp)
print("Matriz de Confusão SVM com LBP:\n", cm_lbp)

# Avaliar o SVM com características PCA
acc_pca, cm_pca = train_evaluate_svm(pca_features, y)
print("Acurácia SVM com PCA:", acc_pca)
print("Matriz de Confusão SVM com PCA:\n", cm_pca)
```

```python
from classifiers import *

# Avaliar o KNN com características HOG
acc_hog_knn, cm_hog_knn = train_evaluate_knn(hog_features, y)
print("Acurácia KNN com HOG:", acc_hog_knn)
print("Matriz de Confusão KNN com HOG:\n", cm_hog_knn)

# Avaliar o KNN com características LBP
acc_lbp_knn, cm_lbp_knn = train_evaluate_knn(lbp_features, y)
print("Acurácia KNN com LBP:", acc_lbp_knn)
print("Matriz de Confusão KNN com LBP:\n", cm_lbp_knn)

# Avaliar o KNN com características PCA
acc_pca_knn, cm_pca_knn = train_evaluate_knn(pca_features, y)
print("Acurácia KNN com PCA:", acc_pca_knn)
print("Matriz de Confusão KNN com PCA:\n", cm_pca_knn)
```

```python
from classifiers import *
from tensorflow.keras.utils import to_categorical
import tensorflow as tf
import numpy as np
# Preparar os dados para a CNN
X_cnn = X.reshape(-1, 35, 35, 1)
y_cnn = to_categorical(y, num_classes=10)

# Dividir os dados em treino e teste
X_train, X_test, y_train, y_test = train_test_split(X_cnn, y_cnn, test_size=0.2, random_state=42)

# Treinar e avaliar a CNN
cnn = build_cnn(input_shape=(35, 35, 1))
cnn.fit(X_train, y_train, epochs=10, batch_size=32, validation_split=0.2)
loss, acc = cnn.evaluate(X_test, y_test)
y_pred = cnn.predict(X_test)
y_pred_classes = np.argmax(y_pred, axis=1)
y_true = np.argmax(y_test, axis=1)
cm_cnn = confusion_matrix(y_true, y_pred_classes)

print("Acurácia CNN:", acc)
print("Matriz de Confusão CNN:\n", cm_cnn)
```
